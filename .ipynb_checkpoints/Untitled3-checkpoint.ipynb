```python
# to implement GRU(Gated recurrent Unit)
# GRU  was introduced to solve (vanishing gradient problem and long term dependency issue)
#Gru can 
# remenber important information
#forget unneccessary information
#train faster than LSTM
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
# sample input data
#shape = (samples, time_steps, features)
X = np.array([
    [[1],[2],[3]],
    [[2],[3],[4]],
    [[3],[4],[5]],
    [[4],[5],[6]]
])

#output labels

Y = np.array([4, 5, 6, 7])

#Build GRU model 

model = Sequential()
#GRU Layer 
model.add(
    GRU(
        units=64,
        activation='tanh',
        input_shape=(3,1)
    ))
model.add(Dense(1))

#compile Model

model.compile(
    optimizer ='adam',
    loss = 'mse'
)

```

```python
#Train model
history = model.fit(
    X, 
    Y,
    epochs=200,
    verbose=1
)
```

```python
# prediction 
test_input = np.array([[[5], [6], [7]]])
prediction = model.predict(test_input)

print("Predicted Output:", prediction)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.show()
```

```python
actual = Y
predicted = model.predict(X)





#prediction visualization
plt.plot(actual, label='Actual')
plt.plot(predicted, label='Predicted')
plt.legend()
plt.show()
```
