```python
#feed forward network for prediction of logic gates 
#prediction of and gate 
import numpy as np
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
#Training Data for and gate
x= np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1] ])
y= np.array([
    [0],
    [0],
    [0],
    [1]
])
```

```python
#create feed neural Network 
model = Sequential()
```

```python
#add hidden layers 
model.add (Dense(4,
                 input_dim=2,
                 activation='sigmoid'))
```

```python
#Output layer 
model.add(Dense(1,
                activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(x,y,
          epochs=1000,
          verbose=0)
```

```python
#prediction 
pred = model.predict(x)
print("Predictions:")
for i in range(len(x)):
    print(x[i], "=>", round(pred[i][0]))
```

```python
#XOR GATE

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# XOR Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

model = Sequential()

model.add(Dense(8,
                input_dim=2,
                activation='relu'))

model.add(Dense(1,
                activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(X, y,
          epochs=2000,
          verbose=0)

pred = model.predict(X)

print("\nXOR Predictions:")
for i in range(len(X)):
    print(X[i], "=>", round(pred[i][0]))
```

```python

# Plot points
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i, 0], X[i, 1],
                    marker='o',
                    s=200,
                    label='Class 0' if i == 0 else "")
    else:
        plt.scatter(X[i, 0], X[i, 1],
                    marker='*',
                    s=300,
                    label='Class 1' if i == 1 else "")

# Labels
plt.xlabel("Input A")
plt.ylabel("Input B")
plt.title("XOR Gate Visualization")
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.grid(True)
plt.legend()

# Show coordinates
for i in range(len(X)):
    plt.annotate(f"({X[i,0]},{X[i,1]})",
                 (X[i,0], X[i,1]),
                 textcoords="offset points",
                 xytext=(5,5))

plt.show()
```
