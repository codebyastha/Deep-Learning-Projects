```python
import pyttsx3# offline tts library
#initialize engine
engine = pyttsx3. init()#create speech engine object
# text to speak
voices = engine.getProperty('voices')
engine.setProperty('voice' , voices[0].id)
text= " Hello Astha"
engine.say(text)
engine.runAndWait()
# convert text to speech 
# convert text into speech
# run the speech engine
#run the speech engine and produces audio output
# Restart Kernel afterevery failed execution:
```
