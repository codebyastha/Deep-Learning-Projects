```python
if TORCH_AVAILABLE:
    class DoubleConv(nn.Module):
        #two consecutive conv->BN->RELU block
        def __init__(self, in_ch, out_ch):
            super().__init__()
            self.block=nn.Sequential(
                nn.Conv2d(in_ch, out_ch, 3, padding=1, bias=False),
                nn.BatchNorm2d(out_ch),
                nn.ReLU(inplace=True),
                nn.Con2d(out_ch, out_ch, 3, padding=1, bias=False),
                nn.BatchNorm2d(out_ch),
                nn.ReLU(inplace=True),
            )
        def forward(self, x):
            return self.block(x)
    class Unet(nn.Module):
        """
        Classic U-Net for binary image segmentation.
        Input:  (B, in_channels, H, W)
        Output: (B, num_classes, H, W)  – raw logits
        """
        def __init__(self, in_channels=1, num_classes=1,features=(64,128,256,512)):
            super().__init__()
            self.downs = nn.ModuleList()
            self.ups = nn.ModuleList()
            self.pool = nn.MaxPool2d(2,2)

            #Encoder
            ch= in_channels
            for f in features:
                self.downs.append(DoubleConv(ch,f))
                ch=f

            #Bottleneck 
            self.bottleneck = DoubleConv(features[-1], features[-1]*2)

            #Decoder

            for f in reversed(features):
                self.ups.append(nn.ConvTranspose2d(f * 2,f,2,2))
                self.ups.appemds(DoubleConv(f*2,f))
            self.final = nn.Conv2d(features[0],num_classes,1)

        

            
                
            
```

```python
 def forward(self, x):
            skips = []
            for down in self.downs:
                x = down(x)
                skips.append(x)
                x = self.pool(x)
 
            x = self.bottleneck(x)
            skips = skips[::-1]
 
            for i in range(0, len(self.ups), 2):
                x = self.ups[i](x)
                skip = skips[i // 2]
                if x.shape != skip.shape:
                    x = F.interpolate(x, size=skip.shape[2:])
                x = torch.cat([skip, x], dim=1)
                x = self.ups[i + 1](x)
 
            return self.final(x)
```

```python
from torch.utils.data import Dataset

data_path = "deep learning/Data"
class SyntheticSegDataset(Dataset):
        """
        Generates random ellipse blobs as foreground objects
        on a noisy background – a clean proxy for medical imaging tasks.
        """
        def __init__(self, n_samples=200, img_size=128, seed=42):
            rng = np.random.default_rng(seed)
            self.images = []
            self.masks  = []
 
            for _ in range(n_samples):
                img  = rng.random((img_size, img_size)).astype(np.float32) * 0.3
                mask = np.zeros((img_size, img_size), dtype=np.float32)
 
                # Add 1–3 elliptical blobs
                n_blobs = rng.integers(1, 4)
                for _ in range(n_blobs):
                    cy, cx = rng.integers(20, img_size - 20, size=2)
                    ry, rx = rng.integers(8, 25, size=2)
                    y, x   = np.ogrid[:img_size, :img_size]
                    ellipse = ((y - cy) / ry) ** 2 + ((x - cx) / rx) ** 2 <= 1
                    img[ellipse]  += rng.random() * 0.5 + 0.4
                    mask[ellipse]  = 1.0
 
                img = np.clip(img, 0, 1)
                self.images.append(img[None])   # (1, H, W)
                self.masks.append(mask[None])   # (1, H, W)
        def __len__(self):
            return len(self.images)
 
        def __getitem__(self, idx):
            return (
                torch.from_numpy(self.images[idx]),
                torch.from_numpy(self.masks[idx]),
            )
```

```python
    # 4. Training Loop
    # ─────────────────────────────────────────
    def dice_loss(pred, target, smooth=1.0):
        pred   = torch.sigmoid(pred)
        num    = (pred * target).sum(dim=(2, 3))
        den    = pred.sum(dim=(2, 3)) + target.sum(dim=(2, 3))
        return 1 - (2 * num + smooth) / (den + smooth)
 
    def combined_loss(pred, target):
        bce  = F.binary_cross_entropy_with_logits(pred, target)
        dice = dice_loss(pred, target).mean()
        return bce + dice
 
    def train_unet(epochs=10, batch_size=16, lr=1e-3, img_size=128):
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print(f"\n── Training U-Net on {device} ──")
 
        train_ds = SyntheticSegDataset(n_samples=200, img_size=img_size, seed=0)
        val_ds   = SyntheticSegDataset(n_samples=50,  img_size=img_size, seed=99)
        train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
        val_dl   = DataLoader(val_ds,   batch_size=batch_size)
 
        model = UNet(in_channels=1, num_classes=1).to(device)
        opt   = optim.Adam(model.parameters(), lr=lr)
        sched = optim.lr_scheduler.CosineAnnealingLR(opt, T_max=epochs)
 
        history = {"train_loss": [], "val_loss": [], "val_iou": []}
 
        for ep in range(1, epochs + 1):
            # ── Train ──
            model.train()
            t_loss = 0
            for imgs, masks in train_dl:
                imgs, masks = imgs.to(device), masks.to(device)
                opt.zero_grad()
                preds = model(imgs)
                loss  = combined_loss(preds, masks)
                loss.backward()
                opt.step()
                t_loss += loss.item()
            sched.step()
            t_loss /= len(train_dl)
```

```python
# 5. Simulation Fallback (no PyTorch)
# ─────────────────────────────────────────────
def simulate_predictions(n=10000, seed=7):
    """
    Simulate realistic segmentation probabilities for
    ROC / PR demo without any deep learning framework.
    """
    rng = np.random.default_rng(seed)
    labels = rng.integers(0, 2, n)
    # Model scores: foreground pixels score higher on average
    probs  = np.where(
        labels == 1,
        rng.beta(5, 2, n),   # foreground: skewed high
        rng.beta(2, 5, n),   # background: skewed low
    )
    return probs, labels
```
