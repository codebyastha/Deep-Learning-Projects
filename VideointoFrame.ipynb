```python
import cv2
import os

#path of the video
video_path= "deep learning/vijay.mp4"
#output_folder
output_folder = "frames"# creates folder to store frames

os.makedirs(output_folder, exist_ok= True)#create folder to store frames

#read video
cap = cv2.VideoCapture(video_path)#open the video
frame_count = 0
while True:
    #read frame 
    success, frame = cap.read() # success = True ->frame exist 
    # frame -> image matrix
    
    
    #stop if video ends
    if not success:
        break
    #frame filename
    frame_name = os.path.join(output_folder,
                           f"frame_{frame_count}.jpg")
    #save frames
    cv2.imwrite(frame_name, frame) # stores frames as image
    print(f"Saved: {frame_name}")
    frame_count+=1
#release the video object
cap.release()  # frees memory resources
print(" ALL frames extracted successfully")
```
