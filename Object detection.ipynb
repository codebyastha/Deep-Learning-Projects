```python
#Write the object detection  using image labeling tool
import torch
import torchvision
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# 1. Instantiate Faster R-CNN Architecture
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(
    pretrained=True
).eval()

# 2. Target Test Asset Location image configuration
img = Image.open("segment_test.jpg").convert("RGB")
transform = torchvision.transforms.ToTensor()
img_tensor = transform(img).unsqueeze(0)

# 3. Model Inference Run
with torch.no_grad():
    predictions = model(img_tensor)[0]

# 4. Process predictions with scores > 0.80
draw = ImageDraw.Draw(img)

for box, score, label in zip(
    predictions['boxes'],
    predictions['scores'],
    predictions['labels']
):
    if score > 0.80:
        box_coords = box.tolist()
        draw.rectangle(box_coords, outline="red", width=3)

# 5. Render Output
plt.figure(figsize=(8, 6))
plt.imshow(img)
plt.axis('off')
plt.show()
```
